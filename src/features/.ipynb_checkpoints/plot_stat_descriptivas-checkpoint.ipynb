```python
import glob 
import os
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def getPathCSV (mfolder):
    # get dir path of script 
    mpath = os.getcwd()
    # get dir folder before to path scripts 
    pathparent = str(Path(mpath).parents[0])
    return pathparent + '/dados/' + mfolder
```

```python
listaNameBacias = [
    '741','7421','7422','744','745','746','7492','751','752','753','754',
    '755','756','757','758','759','7621','7622','763','764','765','766',
    '767','771','772','773', '7741','7742','775','776','777','778','76111',
    '76116','7612','7614','7615','7616','7617','7618','7619', '7613'
]
```

```python
lstFolder =["statsMosaic/", "statsMosaicOld/"]
statFolders = getPathCSV(lstFolder[0])
statFoldersOld = getPathCSV(lstFolder[1])
print("estatisticas por bacias ", statFolders)
print("estatisticas por carta ", statFoldersOld)
yearSel = 2020
baciaSel = listaNameBacias[2]
```

```python
lstCol = [
    'blue_median_mean', 'blue_median_stdDev', 'green_median_mean', 'green_median_stdDev', 
    'red_median_mean','red_median_stdDev', 'nir_median_mean', 'nir_median_stdDev',
    'swir1_median_mean', 'swir1_median_stdDev', 'swir2_median_mean', 'swir2_median_stdDev', 
    'blue_median_wet_mean', 'blue_median_wet_stdDev', 'green_median_wet_mean', 'green_median_wet_stdDev',
    'red_median_wet_mean', 'red_median_wet_stdDev', 'nir_median_wet_mean', 'nir_median_wet_stdDev', 
    'swir1_median_wet_mean', 'swir1_median_wet_stdDev', 'swir2_median_wet_mean', 'swir2_median_wet_stdDev', 
    'blue_median_dry_mean', 'blue_median_dry_stdDev', 'green_median_dry_mean', 'green_median_dry_stdDev', 
    'red_median_dry_mean', 'red_median_dry_stdDev', 'nir_median_dry_mean', 'nir_median_dry_stdDev',
    'swir1_median_dry_mean', 'swir1_median_dry_stdDev', 'swir2_median_dry_mean', 'swir2_median_dry_stdDev'
]
dictBnd = {
    "blue": {"mean": [], "stdDev": []}, 
    "green": {"mean": [], "stdDev": []}, 
    "red": {"mean": [], "stdDev": []}, 
    "nir": {"mean": [], "stdDev": []}, 
    "swir1": {"mean": [], "stdDev": []}, 
    "swir2": {"mean": [], "stdDev": []}, 
}
for kkey in dictBnd.keys():
    print("banda ", kkey)
    for stKey in dictBnd[kkey].keys():
        dictBnd[kkey][stKey] = [kk for kk in lstCol if kkey in kk and stKey in kk]

for kkey, lstB in dictBnd.items():
    print(kkey, lstB)
```

## functions and loading tables 

```python
showDes = False
lstDFstat = []
lstYears = [kk for kk in range(1985, 2024)]
lstfiles = glob.glob(statFolders + "*.csv")
lstfilesOld = glob.glob(statFoldersOld + "*.csv")

for cc, nfile in enumerate(lstfiles):
    #if baciaSel  in nfile:
    print(nfile)
    dftmp = pd.read_csv(nfile)
    dftmp['year'] = lstYears
    dftmp["dados"] = ["statC9"] * dftmp.shape[0]
    # dftmp = dftmp[(dftmp['year'] == yearSel) ]  # &  (dftmp['bacia'] == baciaSel) 
    if showDes:
        print("columns ", dftmp.columns)
        print("shp of table ", dftmp.shape)
        print(dftmp.head())
        print(dftmp["year"].unique())
    lstDFstat.append(dftmp)

for cc, nfile in enumerate(lstfilesOld):
    print("# ", cc, " <> ", nfile)
    dftmp = pd.read_csv(nfile)
    dftmp['year'] = lstYears
    dftmp["dados"] = ["statC8"] * dftmp.shape[0]
    if showDes:
        print("columns ", dftmp.columns)
        print("shp of table ", dftmp.shape)
        print(dftmp.head())
        print(dftmp["year"].unique())
    lstDFstat.append(dftmp)

dfStat  = pd.concat(lstDFstat, axis=0, ignore_index=True)
print("temos {} filas ".format(dfStat.shape))
dfStat.head()
```

```python
print("counts número de dados ", dfStat["dados"].value_counts())
print("size tables joins ", dfStat.shape)
```
